# Analisis inicial de dataset

```python
import os, shutil

baseDir = './images'
i=0
l=0
minSize=1000
maxSize=0

for x in os.listdir(baseDir):
    subDir = baseDir + "/" + x
    size = os.listdir(subDir)
    i=i+1
    l = len(size)
    if l < minSize:
        minSize = l
    if l > maxSize:
        maxSize = l
        
print("maxSize ", maxSize)
print("minSize ", minSize)
```

# Preparando los datos

```python
# Directorio de trabajo
processDir = './process'
os.makedirs(processDir, exist_ok=True)

# Directorios para entrenamiento, test y validacion
train_dir = os.path.join(processDir, 'train')
os.makedirs(train_dir, exist_ok=True)

validation_dir = os.path.join(processDir, 'validation')
os.makedirs(validation_dir, exist_ok=True)

test_dir = os.path.join(processDir, 'test')
os.makedirs(test_dir, exist_ok=True)
```

```python
#Porcentajes para el split de datos
porTest = 15
porVal = 15

for clase in os.listdir(baseDir):
    subDir = baseDir + "/" + clase
    files = os.listdir(subDir)
    
    # Crea el directorio en test
    categ_test_dir = os.path.join(test_dir, clase)
    os.makedirs(categ_test_dir, exist_ok=True)
    
    cantTest = (len(files) * porTest) // 100
    
    # Copiamos las imagenes test
    fnames = files[0:cantTest]
    for fname in fnames:
        src = os.path.join(subDir, fname)
        dst = os.path.join(categ_test_dir, fname)
        shutil.copyfile(src, dst)

    # Crea el directorio en validacion
    categ_val_dir = os.path.join(validation_dir, clase)
    os.makedirs(categ_val_dir, exist_ok=True)
    
    cantVal = (len(files) * porVal) // 100

    # Copiamos las imagenes validacion
    fnames = files[cantTest: (cantTest + cantVal)]
    for fname in fnames:
        src = os.path.join(subDir, fname)
        dst = os.path.join(categ_val_dir, fname)
        shutil.copyfile(src, dst)

    # Crea el directorio en train
    categ_train_dir = os.path.join(train_dir, clase)
    os.makedirs(categ_train_dir, exist_ok=True)

    cantTrain = len(files) - (cantTest + cantVal)

    # Copiamos las imagenes train
    fnames = files[(cantTest + cantVal): len(files)]
    for fname in fnames:
        src = os.path.join(subDir, fname)
        dst = os.path.join(categ_train_dir, fname)
        shutil.copyfile(src, dst)
    
    print("dir ", subDir, len(files))
    print("categ_test_dir  ", categ_test_dir,cantTest)
    print("categ_val_dir   ", categ_val_dir, cantVal)
    print("categ_train_dir ", categ_train_dir , cantTrain)
```
